# Exercici 1
Una variable numèrica.

```python
import matplotlib.pyplot as plt

# Seleccionar una variable numérica (por ejemplo, el monto de las transacciones)
variable_numerica = 'amount'

# Graficar el histograma
plt.hist(df_transactions[variable_numerica], bins=20, edgecolor='white', color ='mediumaquamarine')
plt.title(f'Histograma de {variable_numerica}')
plt.xlabel(variable_numerica)
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Hacemos un merge entre las tablas para establecer las relaciones
df_merged = pd.merge(df_transactions, df_transaction_products, left_on='id', right_on='transaction_id')
df_merged = pd.merge(df_merged, df_products, left_on='product_id', right_on='id')
```

```python
# Seleccionar dos variables numéricas (por ejemplo, latitud y longitud de las transacciones)
var_num1 = 'price_usd'
var_num2 = 'amount'

# Graficar el scatter plot
plt.scatter(df_merged[var_num1], df_merged[var_num2])
plt.title(f'Relación entre {var_num1} y {var_num2}')
plt.xlabel(var_num1)
plt.ylabel(var_num2)
plt.show()
```
